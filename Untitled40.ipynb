```python
import bar_chart_race as bcr
import pandas as pd
import os

#os.chdir('corona_dat.csv')
data_corona=pd.read_csv("corona_dat.csv")
data_corona.head()

cols=['date','Italy','China','Australia','Brazil','India','United Kingdom','Pakistan','Spain','New Zealand','Russia','Thailand']
subsetdf=data_corona[cols]
subsetdf.set_index("date",inplace=True)
cum_sum_df=subsetdf.cumsum(axis=0)
bcr.bar_chart_race(df=cum_sum_df,filename=None,figsize=(3.5,3),title='covid_19 report')
```
